```python
import pandas as pd
import numpy as np
train=pd.read_csv('C:/Users/ADHI/Desktop/MACHINE LEARNING/train.csv')
test=pd.read_csv('C:/Users/ADHI/Desktop/MACHINE LEARNING/test.csv')
train=train.drop(['Name','Ticket','Fare','Cabin','Embarked'],axis=1)
train.head()
```

```python

from sklearn.preprocessing import Imputer
Imp=Imputer(missing_values='NaN',strategy='median',axis=1)
new=Imp.fit_transform(train.Age.values.reshape(1,-1))
train['Age1']=new.T
train.drop('Age',axis=1,inplace=True)
train.isnull().sum()
```

```python
X = train.iloc[:,2:7].values
y = train.iloc[:, 1].values
```

```python
X
```

```python
# Encoding categorical data
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
labelencoder = LabelEncoder()
X[:, 1] = labelencoder.fit_transform(X[:, 1])
```

```python
X[:, 1]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25, random_state = 0)
```

```python
from sklearn import preprocessing
sc = preprocessing.RobustScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = classifier.predict(X_test)
print (y_pred)
```

```python
from sklearn import metrics
metrics.accuracy_score(y_pred, y_test)
```
